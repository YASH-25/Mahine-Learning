```python
import pandas as pd 

data = pd.read_csv("Drug_Consumption_Quantified.csv")
data.head()
```

**Encode the String values into Integer**

```python
stringValues = ['Ethnicity','Gender',"Age", "Education", "Country", "Nscore", "Escore",'Alcohol','Amphet','Amyl','Benzos','Caff','Cannabis','Choc','Coke','Crack','Ecstasy','Heroin','Ketamine','Legalh','LSD','Meth','Mushrooms','Nicotine','Semer','VSA']
for i in stringValues:
  # get all unique values in a list. Index of each 
  uniqueList = list(set(data[i]))
  # Create replace dictionary with key as string to be replaced and value as integer encoding for the string
  replaceDict = {}
  for j in range(len(uniqueList)):
    replaceDict[uniqueList[j]] = j
  data = data.replace({i : replaceDict})

data.head()
```

** Identify the Unique values**

```python
for i in data.columns:
  print(i ,list(set(data[i])))
```

**Find Missing values**

```python
data.isnull()
```

**Replace Missing Values**

```python
for i in data.columns:
  if(data[i].isnull().values.any()):
    data[i].fillna(data[i].mean(),inplace = True)
    print("Replaced null values with average in the column, ", i)
  else:
    print("No null values for the column :", i)
```

**Normalizing the data using standard scaler or minmax scaler**

```python
x_cols = ['ID', 'Age', 'Gender', 'Education', 'Country', 'Ethnicity', 'Nscore',
       'Escore', 'Oscore', 'AScore', 'Cscore', 'Impulsive', 'SS']
```

```python
for i in x_cols:
    data[i] = data[i]  / data[i].abs().max()
# view normalized data
display(data)
```

**Apply KNN for the preprocessed dataset use 5 different K Values and 3 different distance measures**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
# split the data set
numRows = data.shape[0]
trainSize = round(numRows * 0.8) 
train = data.iloc[:trainSize, : ]
test = data.iloc[trainSize : , :]
y_col = ["Ecstasy"]

knn.fit(train[x_cols], train[y_col])
```

```python
train[y_col].head()
```

```python
train[x_cols].head()
```

**Display the confusion matrix, accuracy measures, classification report.**

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
plot_confusion_matrix(knn, test[x_cols], test[y_col])
```

**Trainig Accuracy**

```python
knn.score(train[x_cols], train[y_col])
```

**Testing Accuracy**

```python
knn.score(test[x_cols], test[y_col])
```

**Relationship between Accuracy vs k value**

 metric = minkowski, p = 3

```python
import matplotlib.pyplot as plt
x = []
trainY = []
testY = []
for i in range(1,200):
  x.append(i)
  knn = KNeighborsClassifier(n_neighbors=i, p = 3, metric = "minkowski")
  knn.fit(train[x_cols], train[y_col])
  trainY.append(knn.score(train[x_cols], train[y_col]))
  testY.append(knn.score(test[x_cols], test[y_col]))
```

```python
maxY = max(testY)
maxIndex = testY.index(maxY)
maxX = x[maxIndex]
print("Maximum accuracy : ", maxY)
plt.plot(x, testY, label = "Test Accuracy")
plt.plot(x,trainY , label = "Train accuracy")
plt.xlabel("K-value")
plt.ylabel("Accuracy")
plt.annotate("k = " + str(maxX), xy=(maxX, maxY))
plt.legend()
plt.show()
```
