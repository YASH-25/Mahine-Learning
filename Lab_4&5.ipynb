```python
import numpy as np
import pandas as pd
import sklearn as sns
from sklearn.cluster import  KMeans
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("Drug_Consumption.csv",header=0)
```

```python
x_train,x_test=train_test_split(df,test_size=0.2)
```

```python
km= KMeans(n_clusters=4)
km
```

```python
stringValues = ['Ethnicity','Gender',"Age", "Education", "Country", "Nscore", "Escore",'Alcohol','Amphet','Amyl','Benzos','Caff','Cannabis','Choc','Coke','Crack','Ecstasy','Heroin','Ketamine','Legalh','LSD','Meth','Mushrooms','Nicotine','Semer','VSA']
for i in stringValues:
  # get all unique values in a list. Index of each 
  uniqueList = list(set(df[i]))
  # Create replace dictionary with key as string to be replaced and value as integer encoding for the string
  replaceDict = {}
  for j in range(len(uniqueList)):
    replaceDict[uniqueList[j]] = j
  df = df.replace({i : replaceDict})

df.head()
```

```python
y_predicted=km.fit_predict(df[['Age','Country']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
df5=df[df.cluster==4]

plt.scatter(df1.Age,df1['Nscore'],color='green')
plt.scatter(df2.Age,df2['Nscore'],color='red')
plt.scatter(df3.Age,df3['Nscore'],color='black')
plt.scatter(df4.Age,df4['Nscore'],color='yellow')
plt.scatter(df5.Age,df5['Nscore'],color='blue')


plt.xlabel('Age')
plt.ylabel('Nscore')
plt.legend()
plt.show()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]


plt.scatter(df1.Age,df1['Escore'],color='green')
plt.scatter(df2.Age,df2['Escore'],color='red')
plt.scatter(df3.Age,df3['Escore'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="blue",marker="*",label="centroid")

plt.xlabel('Age')
plt.ylabel('Escore')
plt.legend()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv("/content/Drug_Consumption.csv")
```

```python
data.columns
```

```python
strings = ["Ecstasy"]
for i in strings:
  print(i)
  uniqueList = data[i].unique()
  dic = {}
  for i in range(len(uniqueList)):
    dic[uniqueList[i]] = i
  data.replace({ i : dic }, inplace = True)
```

```python
data.head()
```

```python
data = data.replace({"Ecstasy" : dic})
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

x = data["Ecstasy"]
y = data["Nscore"]
```

```python
data = list(zip(x, y))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
```
